# Agentic AI Bootcamp
---

## Learning Objectives

1. Understand the architecture and capabilities of NVIDIA Inference Microservices (NIMs) and how to deploy them locally for accelerated AI inference.
2. Understand the Model Context Protocol (MCP) and how it enables standardized communication between AI models and external tools/data sources.
3. Learn to use LangGraph for designing stateful agent graphs that coordinate multi-step reasoning, tool invocation, and decision-making logic while leveraging NVIDIA NIM for accelerated inference.
4. Gain hands-on experience with the NeMo Agent Toolkit by deploying Nemotron-3-Nano as a local NIM endpoint and configuring ReAct agents with external tool integrations (such as Wikipedia search) using declarative workflow definitions
5. Apply learned concepts in a challenge exercise to build a complete NIM-powered agentic application.

### Tutorial Duration
The total lab session is 8 hours, consisting of a 2-hour material presentation, a 2-hour tutorial, and a 4-hour challenge.

### Content Level
Beginner, Intermediate

### Target Audience and Prerequisites
The target audience for these labs are researchers, developers, and graduate students who are interested in Retrieval Augmented Generation and NIM approach to solving Generative AI tasks via the use of GPUs. Audiences are expected to have Python programming and Natural Language Processing Knowledge and possess NVIDIA NGC and API keys.
