## MCP high level SDK

```python
%%writefile mcp_server/simple_server.py

from fastmcp import FastMCP

mcp = FastMCP("My MCP Server", stateless_http=True)

@mcp.tool
def greet(name: str) -> str:
    return f"Hello, {name}!"

@mcp.tool
def add(a: int, b: int) -> int:
    """Add two numbers"""
    return a + b
```
